```python
electricity_end_use = electricity_end_use.groupby('Date')['Value'].sum().reset_index()
```

```python
# Trend
# Convert 'Date' column to datetime if it's not already in datetime format
electricity_end_use['Date'] = pd.to_datetime(electricity_end_use['Date'])

# Plot the trend
plt.figure(figsize=(15, 8))  # Adjust the figsize to make the graph larger
plt.plot(electricity_end_use['Date'], electricity_end_use['Value'], marker='o', linestyle='-')
plt.title('Electricity Demand Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)

# Customize x-axis tick marks and labels to show year with 5-year gap
plt.xticks(pd.date_range(start=electricity_end_use['Date'].min(), end=electricity_end_use['Date'].max(), freq='5Y'))

plt.show()
```

```python
# Build and fit the ARIMA model

# Convert 'Date' column to datetime
electricity_end_use['Date'] = pd.to_datetime(electricity_end_use['Date'])

# Set 'Date' column as index
electricity_end_use.set_index('Date', inplace=True)

# Split data into train and test sets
train_size = int(len(electricity_end_use) * 0.8)
train, test = electricity_end_use.iloc[:train_size], electricity_end_use.iloc[train_size:]

# Grid search for ARIMA parameters
best_rmse = np.inf
best_params = None

for p in range(3):  # Range for p
    for d in range(3):  # Range for d
        for q in range(3):  # Range for q
            try:
                model_arima = ARIMA(train, order=(p, d, q))
                model_arima_fit = model_arima.fit()
                predictions_arima = model_arima_fit.predict(start=len(train), end=len(train)+len(test)-1, typ='levels')
                rmse = np.sqrt(mean_squared_error(test['Value'], predictions_arima))
                
                if rmse < best_rmse:
                    best_rmse = rmse
                    best_params = (p, d, q)
                    
                print(f"ARIMA({p},{d},{q}) - RMSE: {rmse}")
            except:
                continue

print(f"Best ARIMA Parameters: {best_params} - Best RMSE: {best_rmse}")

# Fit the ARIMA model with the best parameters
model_arima_best = ARIMA(train, order=best_params)
model_arima_fit_best = model_arima_best.fit()
predictions_arima_best = model_arima_fit_best.predict(start=len(train), end=len(train)+len(test)-1, typ='levels')

# Plot actual vs. predicted values
plt.figure(figsize=(12, 6))
plt.plot(test.index, test['Value'], label='Actual')
plt.plot(test.index, predictions_arima_best, label='Predicted (ARIMA)', color='blue')
plt.title('Actual vs. Predicted Electricity End Use (ARIMA)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

# Calculate RMSE for ARIMA model with the best parameters
rmse_arima_best = np.sqrt(mean_squared_error(test['Value'], predictions_arima_best))

# Calculate the error percentage and accuracy for ARIMA model with the best parameters
value_range = test['Value'].max() - test['Value'].min()
error_percentage_arima_best = (rmse_arima_best / value_range) * 100
accuracy_arima_best = 100 - error_percentage_arima_best

print('Error Percentage (ARIMA model with best parameters):', error_percentage_arima_best, '%')
print('Accuracy (ARIMA model with best parameters):', accuracy_arima_best, '%')
```

```python
# Build and fit the SARIMA model

# Split data into train and test sets
train_size = int(len(electricity_end_use) * 0.8)
train, test = electricity_end_use.iloc[:train_size], electricity_end_use.iloc[train_size:]

# Build and fit the SARIMA model
model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
model_fit = model.fit()

# Make predictions
predictions = model_fit.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)

# Plot actual vs. predicted values
plt.figure(figsize=(12, 6))
plt.plot(test.index, test['Value'], label='Actual')
plt.plot(test.index, predictions, label='Predicted', color='red')
plt.title('Actual vs. Predicted Electricity End Use')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(test['Value'], predictions))

# Calculate the range of values in the test set
value_range = test['Value'].max() - test['Value'].min()

# Calculate the error percentage
error_percentage = (rmse / value_range) * 100

# Calculate accuracy
accuracy = 100 - error_percentage

print('Error Percentage:', error_percentage, '%')
print('Accuracy:', accuracy, '%')
```

```python
# Build and fit the Holt-Winters model
model_hw = ExponentialSmoothing(train, seasonal='add', seasonal_periods=12)
model_hw_fit = model_hw.fit()

# Make predictions
predictions_hw = model_hw_fit.predict(start=len(train), end=len(train)+len(test)-1)

# Plot actual vs. predicted values
plt.figure(figsize=(12, 6))
plt.plot(test.index, test['Value'], label='Actual')
plt.plot(test.index, predictions_hw, label='Predicted (Holt-Winters)', color='green')
plt.title('Actual vs. Predicted Electricity End Use (Holt-Winters)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

# Calculate RMSE for Holt-Winters model
rmse_hw = np.sqrt(mean_squared_error(test['Value'], predictions_hw))

# Calculate the error percentage and accuracy for Holt-Winters model
error_percentage_hw = (rmse_hw / value_range) * 100
accuracy_hw = 100 - error_percentage_hw

print('Error Percentage (Holt-Winters model):', error_percentage_hw, '%')
print('Accuracy (Holt-Winters model):', accuracy_hw, '%')
```

```python
# Forecast

# Fit Holt-Winters model to the entire dataset
model = ExponentialSmoothing(electricity_end_use, trend="add", seasonal="add", seasonal_periods=12)
hw_model = model.fit()

# Forecast for the next 10 years
forecast = hw_model.forecast(10*12+1)

forecast = forecast.iloc[1:]

# Aggregate forecasted values into yearly sums
forecast_yearly = forecast.resample('Y').sum()

# Extract only the year from the index
forecast_yearly.index = forecast_yearly.index.year

# Plot forecast
plt.figure(figsize=(12, 6))
plt.plot(electricity_end_use.index, electricity_end_use, label='Original Data')
plt.plot(forecast.index, forecast, label='Forecast')
plt.title('Electricity Demand Forecast')
plt.xlabel('Year')
plt.ylabel('Demand')
plt.legend()
plt.grid(True)
plt.show()


# Print forecasted yearly sums
print("Forecasted yearly for the next 10 years:")
print(forecast_yearly)
```
